```python
import numpy as np
import matplotlib.pyplot as plt
import os
import time
import cv2
import csv
import math
from tqdm import tqdm
```

```python
def check_pct(arr):
    nans = np.isnan(arr)
    zeros = (arr == 0)
    neg_outlier = (arr < 0)
    pos_outlier = (arr>20)
    assert np.sum(pos_outlier) <1
    count = np.sum(nans) + np.sum(zeros) + np.sum(neg_outlier) +np.sum(pos_outlier)
    #loss rate
    pct = count/(256*256) *10 
    temp_pct = pct*10
    pct = math.floor(pct)*10
    return temp_pct, pct
```

```python
########## path ##########
data_base = '/media/ubuntu/My Book/data/Chl-a/train/perfect'
save_base = '/media/ubuntu/My Book/data/Chl-a/train/perfect_backup'

if not os.path.isdir(save_base):
    os.makedirs(save_base)
##########################
imgs = os.listdir(data_base)
```

```python
cnt=0
s_cnt=0
p_cnt=0
a_cnt=0
v_cnt=0
for i in tqdm(imgs):
    i_path = os.path.join(data_base, i)
    img = cv2.imread(i_path, cv2.IMREAD_UNCHANGED)
    t_pct, _ = check_pct(img)
    if t_pct<.5:
        cnt+=1
    if t_pct<.1:
        s_cnt+=1
    if t_pct<.05:
        p_cnt+=1
    if t_pct<.01:
        a_cnt+=1
    if t_pct<.001:
        v_cnt+=1
        
#         save_path = os.path.join(save_base, i)
#         shutil.move(i_path, save_path)

```
