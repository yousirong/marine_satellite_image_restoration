```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

def load_image(file_path):
    """
    Load PNG image and return as numpy array.
    """
    return cv2.imread(file_path, cv2.IMREAD_UNCHANGED)

def convert_to_raw_chl(image, scale_factor=1):
    """
    Convert the image to raw Chlorophyll-a values.
    This function assumes some scaling factor for conversion.
    Adjust `scale_factor` based on the dataset format.
    """
    chl_raw = image * scale_factor
    return chl_raw

def plot_image_with_unique_values(image, title):
    """
    Display the image and print its unique values.
    """
    unique_vals = np.unique(image)
    print(f"Unique values in {title}: {unique_vals}")
    
    # Display image
    plt.imshow(image, cmap='viridis')
    plt.title(f'{title} (Unique Values)')
    plt.colorbar(label='Chlorophyll-a Concentration (mg/m³)')
    plt.show()

def process_test_images(directory, image_prefix="gt"):
    """
    Process the test images in the provided directory.
    - Load GT, mask, masked, and recon images.
    - Convert them to raw Chlorophyll-a values.
    - Print and display their unique values.
    """
    image_types = ['gt', 'mask', 'masked', 'recon','img']  # Expected image types
    
    for img_type in image_types:
        # Find images of this type in the directory
        img_files = sorted([f for f in os.listdir(directory) if f.startswith(img_type) and f.endswith('.png')])
        print(f"Processing {img_type.upper()} images...")

        for img_file in img_files:
            file_path = os.path.join(directory, img_file)
            print(f"Loading {file_path}...")

            # Load the image
            img = load_image(file_path)

            # Convert to raw Chlorophyll-a values
            chl_img = convert_to_raw_chl(img, scale_factor=1)  # Adjust scale_factor as needed
            
            # Plot and print unique values
            plot_image_with_unique_values(chl_img, title=f'{img_type.upper()} Image')

# Example usage:
test_directory = "/home/juneyonglee/Documents/ust21_8day_results"  # Replace with your actual directory path
process_test_images(test_directory)
```
